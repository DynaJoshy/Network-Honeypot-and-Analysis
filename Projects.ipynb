```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the honeypot logs data
df = pd.read_csv('honeypot_logs.csv', parse_dates=['timestamp'])

# Display the first few rows of the dataframe
print(df.head())

# Plot the number of attacks over time
df.set_index('timestamp', inplace=True)
attack_counts = df.resample('D').size()

plt.figure(figsize=(12, 6))
plt.plot(attack_counts, marker='o')
plt.title('Number of Attacks Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()

# Plot the top 10 source IPs
top_source_ips = df['source_ip'].value_counts().head(10)

plt.figure(figsize=(10, 6))
top_source_ips.plot(kind='bar')
plt.title('Top 10 Source IPs')
plt.xlabel('Source IP')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Number of sample records
num_records = 1000

# Generate random timestamps
end_date = datetime.now()
start_date = end_date - timedelta(days=30)
timestamps = [start_date + timedelta(seconds=random.randint(0, int((end_date - start_date).total_seconds()))) for _ in range(num_records)]

# Generate random source IPs
source_ips = [f"192.168.{random.randint(0, 255)}.{random.randint(0, 255)}" for _ in range(num_records)]

# Generate random destination ports
destination_ports = [random.choice([22, 80, 443, 8080, 3306]) for _ in range(num_records)]

# Generate random attack types
attack_types = [random.choice(['Brute Force', 'DDoS', 'SQL Injection', 'XSS', 'Ransomware']) for _ in range(num_records)]

# Create a DataFrame
df = pd.DataFrame({
    'timestamp': timestamps,
    'source_ip': source_ips,
    'destination_port': destination_ports,
    'attack_type': attack_types
})

# Save to CSV
df.to_csv('honeypot_logs.csv', index=False)

print(df.head())
```
